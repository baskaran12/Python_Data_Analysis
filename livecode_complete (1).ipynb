## Complete scraper that can navigate links!

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

BASE_URL = "https://recipes.lewagon.com/"


home_html   = requests.get(BASE_URL).content
home_soup   = BeautifulSoup(home_html, "html.parser")
recipes     = home_soup.select(".recipe")

records = []

for r in recipes:
    title = r.find("p", class_="recipe-name").get_text(strip=True)
    cook_time=r.find("span",class_="recipe-cooktime").text.strip()
    recipe_url = BASE_URL + r.a["href"]
    #print(recipe_url)
    detail_html = requests.get(recipe_url).content
    detail_soup = BeautifulSoup(detail_html, "html.parser")

    desc_tag = detail_soup.find("p", class_="recipe-description")
    description = desc_tag.get_text(strip=True) if desc_tag else "" 

    all_ingr=detail_soup.find_all("p",class_="mb-0")
    ingr_lst=[]
    for list_ingr in all_ingr:
        ingr_lst.append(list_ingr.text.strip())

    records.append({"title": title, "description": description,"ingredients":ingr_lst})


df = pd.DataFrame(records)

df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

BASE_URL = "https://recipes.lewagon.com/"

# this is where we'll put all our recipes from all pages
all_recipes = []

# let's loop through pages 1, 2, 3, 4, 5
# you can change this number to get more or less pages
for page_number in range(1, 6):  
    
    print(f"hey! scraping page {page_number} now...")
    
    # here wea are building the url for each page
    # page 1 is just the main url
    # page 2 becomes the main url + "?page=2"
    # page 3 becomes the main url + "?page=3" and so on...
    if page_number == 1:
        url = BASE_URL
    else:
        url = BASE_URL + f"?page={page_number}"
    
    # now let's grab the html from this page
    page_html = requests.get(url).content
    page_soup = BeautifulSoup(page_html, "html.parser")
    
    # look for all the recipes on this page
    recipes = page_soup.select(".recipe")
    
    # if there are no recipes on this page, we are done!!!!!!!!!!!!!!!!!
    # this happens when we reach the last page
    if not recipes:
        print("oops, no more recipes found! we're done here")
        break
    
    # now lets go through each recipe on this page
    for recipe in recipes:
        
        # get the recipe title
        title = recipe.find("p", class_="recipe-name").get_text(strip=True)
        
        # build the full url for this specific recipe
        recipe_url = BASE_URL + recipe.a["href"]
        
        # now we need to visit each recipe page to get the description
        # this takes a bit more time but gets us better data
        detail_html = requests.get(recipe_url).content
        detail_soup = BeautifulSoup(detail_html, "html.parser")
        
        # try to find the description, if it doesn't exist just use empty string
        desc_tag = detail_soup.find("p", class_="recipe-description")
        description = desc_tag.get_text(strip=True) if desc_tag else ""
        
        # add this recipe to our big list of all recipes (please read about lists in Python)
        all_recipes.append({
            "title": title,
            "description": description
        })

# finally, let's turn our list into a nice dataframe and see what we got
df = pd.DataFrame(all_recipes)
df.head()
```
